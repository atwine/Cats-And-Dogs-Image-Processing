Let's list the directory structure, the reason we are doing this is because we are going to have to restructure the whole directory structure

```python
# function to list files in each folder of the current working directory

def list_files(startpath):
    for root, dirs, files in os.walk(startpath):
        # print(dirs)
        if dir!= '.git':
            level = root.replace(startpath, '').count(os.sep)
            indent = ' ' * 4 * (level)
            print('{}{}/'.format(indent, os.path.basename(root)))
            subindent = ' ' * 4 * (level + 1)
            for f in files:
                print('{}{}'.format(subindent, f))
```

```python
#for this we will need regular expressions
import re
```

```python
string = ''''
        AtwineMugume_234343
        '''

pattern = re.compile(r'([A-Z][a-z]_?[a-zA-Z_]+[a-zA-Z]+[a-zA-Z]+)')
matches = pattern.findall(string)
for match in matches:
    print(match)
```

```python
#let's create a function to do this so we won't have to rewrite it for train and test
def split_df(df):
    #we have to change the data type so that its readable by regular expressions
    df.astype('str')
    cats =[]
    dogs =[]
    
    #we want to read the value in each cell
    for i in df.index:
    
        val = df.loc[i,'Breed']
        
        #string to search
        
        #pattern for the words
        cat_ptn = re.compile(r'[A-Z][a-z]_?[a-zA-Z_]+[a-zA-Z]+[a-zA-Z]+')
        dog_ptn = re.compile(r'[a-z]_?[a-zA-Z_]+[a-zA-Z]+[a-zA-Z]+')
        
        #matches
        cat_match = cat_ptn.findall(val) 
        dog_match = dog_ptn.findall(val)
        
        #cats
        for match in cat_match:
            cats.append(match)
            
        #cats
        for dog in dog_match:
            dogs.append(dog) 
            

    return dogs, cats
```

```python
dogs, cats = split_df(train)
```
